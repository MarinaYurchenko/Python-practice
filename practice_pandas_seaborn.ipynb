```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df_dia = sns.load_dataset("diamonds")
df_dia.columns # check the columns in the dataset
df_dia.head(100) # sample, tail - to overview dataset 
```

```python
# 1. Ð¯ÐºÐµ ÑÐµÑ€ÐµÐ´Ð½Ñ” Ð·Ð½Ð°Ñ‡ÐµÐ½Ð½Ñ Ñ†Ñ–Ð½Ð¸ Ð²ÑÑ–Ñ… Ð°Ð»Ð¼Ð°Ð·Ñ–Ð² Ñƒ Ð´Ð°Ñ‚Ð°ÑÐµÑ‚Ñ–? - Ð·Ð±ÐµÑ€ÐµÐ³Ñ‚Ð¸ Ð² Ð·Ð¼Ñ–Ð½Ð½Ñƒ avg_df Ñ‚Ð° Ð¾ÐºÑ€ÑƒÐ³Ð»Ð¸Ñ‚Ð¸ Ð´Ð¾ Ð´Ñ€ÑƒÐ³Ð¾Ð³Ð¾ Ð·Ð½Ð°Ñ‡ÐµÐ½Ð½Ñ 
sns.scatterplot(x='price', y='carat', data=df_dia)

avg_price=df_dia['price'].mean()
print(f'The average price of all diamonds in this dataset is: {round(avg_price,2)}')  # Ñ‡Ð¾Ð¼ÑƒÑÑŒ Ð¿Ð¾Ð²ÐµÑ€Ñ‚Ð°Ñ”Ñ‚ÑŒÑÑ Ð»Ð¸ÑˆÐµ Ð¾Ð´Ð¸Ð½ Ð·Ð½Ð°Ðº Ð¿Ñ–ÑÐ»Ñ ÐºÐ¾Ð¼Ð¸
```

```python
# 2. Ð¯ÐºÐ° ÑÐµÑ€ÐµÐ´Ð½Ñ ÐºÑ–Ð»ÑŒÐºÑ–ÑÑ‚ÑŒ ÐºÐ°Ñ€Ð°Ñ‚Ñ–Ð² Ñƒ Ð°Ð»Ð¼Ð°Ð·Ñ–Ð² ÐºÐ¾Ð¶Ð½Ð¾Ñ— ÑÐºÐ¾ÑÑ‚Ñ– Ð¾Ð³Ñ€Ð°Ð½ÑŽÐ²Ð°Ð½Ð½Ñ?
quant_count=df_dia['cut'].value_counts() # general quantity of diamonds in each cut
perc_count=df_dia['cut'].value_counts(normalize= True) * 100 # percentage quantity of diamonds in each cut
print(f'{quant_count} \n\n{perc_count}')

```

```python
#4. Ð¯ÐºÐ¸Ð¹ Ñ€Ð¾Ð·Ð¿Ð¾Ð´Ñ–Ð» ÐºÑ–Ð»ÑŒÐºÐ¾ÑÑ‚Ñ– Ð°Ð»Ð¼Ð°Ð·Ñ–Ð² Ð·Ð° ÐºÐ¾Ð¶Ð½Ð¸Ð¼ ÐºÐ¾Ð»ÑŒÐ¾Ñ€Ð¾Ð¼? Ð’Ð¸ÐºÐ¾Ñ€Ð¸ÑÑ‚Ðµ Ð³Ñ€Ð°Ñ„Ñ–Ðº Ð´Ð»Ñ Ð¿Ð¾ÑÑÐ½ÐµÐ½Ð½Ñ
color_counts = df_dia['color'].value_counts()

plt.figure(figsize=(8, 6))
sns.barplot(x=color_counts.index, y=color_counts.values, palette='pastel')
plt.title('Diamonds by Color')
plt.xlabel('Color')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```
