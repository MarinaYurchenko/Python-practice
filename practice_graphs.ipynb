```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
diamonds = sns.load_dataset("diamonds")
```

```python
# 1. Load the “diamonds” dataset from seaborn.
diamonds.columns
```

```python
# 2. Show the countplot of the ‘cut’ field.
plt.figure(figsize = (6,4)) # встановити розмір графіку 

sns.countplot(x = diamonds.cut ,data = diamonds); # показати кіл-к данних

plt.xlabel('Cut type') # встановити назву по осі x
plt.ylabel('Count') # встановити назву по осі y
plt.title('Cut') # встановити назву графіку
plt.show() 
```

```python
# 3. Show the distplot of the ‘price’ field.
sns.displot(x='price', data = diamonds, hue='cut', kind = 'kde', fill = True, palette= 'magma')
plt.title('Графік розподілу price')
plt.show()
```

```python
# 4. Show the boxplot of the price based on cut. Do the results make sense?
plt.subplots(figsize = (8,6))
sns.boxplot(data = diamonds, x = 'cut', y = 'price', palette = 'inferno');
```

```python
# 5. Show the violin plot of the size (in carat) based on cut. What information does this add to ex. 4?
sns.violinplot(data= diamonds, x = 'cut', y = 'carat', palette = 'inferno');
```

```python
# 6. Show the same plot as ex. 5, this time split by diamond color
sns.violinplot(data= diamonds, x = 'cut', y = 'carat', hue = 'color', split = True);
```

```python
sns.barplot(x= 'cut', y = 'carat' , data = diamonds, estimator= np.mean, palette = 'magma') 
```

```python
# > 1. Use the same diamonds dataset; filter the dataset only to diamonds of size up to 0.6 carat and price up to $2000.
# > 2. join plot between price and size

diamonds_1=diamonds[(diamonds['carat'] <= 0.6) & (diamonds['price'] <= 2000)]
sns.jointplot(data=diamonds_1, x='carat', y='price', kind = 'hex');
```

```python
tips = sns.load_dataset('tips')
tips.head(10)
```

```python
# > 3. Load the “tips” dataset. Show a pairplot of the dataset.
sns.pairplot(tips, hue='sex', diag_kind='kde');
```

```python
# > 4. Add a new column for whether the meal was on a weekend or weekday.
def type_day(day):
    if day in ['Sat', 'Sun']:
        return 'Weekend'
    else:
        return 'Weekday'

tips['day_type']=tips['day'].apply(type_day)
```

```python
tips.sample(50) #checking the result
```

```python

#> 5. Show a pairplot of the dataset split by weekday / weekend.

sns.pairplot(tips, hue='day_type', diag_kind='kde');
```
